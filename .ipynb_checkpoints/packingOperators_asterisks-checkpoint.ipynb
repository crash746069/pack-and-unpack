```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize']=[11,9]
```

```python
fruits = ['pear','lemon','apple','banana']
print(fruits)
print(fruits[0],fruits[1],fruits[2],fruits[3])
print(*fruits)   # expands an iterable as if it were multiple arguments to the function
```

```python
def transpose_list(list_of_lists):
    return [
        list(row)
        for row in zip(*list_of_lists)
    ]

L = [[1,2],[3,4],[5,6]]
M = transpose_list(L)
print(L)
print([row for row in zip([1,2],[3,4],[5,6])])
print(M)

# Again * is unpacking a list as if it were providing three separate arguments to the zip function
```

```python
D = {'linewidth':10,'label':'here is my label'}
plt.plot([-4,-3,-2,-1,0,1,2,3,4],[16,9,4,1,0,1,4,9,16],**D)
```

```python
date_info = {'day': "01", 'month': "01", 'year': "2020"}
track_info = {'artist': "Beethoven", 'title': 'Symphony No 5'}
filename = "{year}-{month}-{day}-{artist}-{title}.txt".format(
    **date_info,
    **track_info,
)
print(filename)
```

```python
print('the quick brown %s jumps over the lazy %s' % ('fox','dog'))

print('the quick brown {} jumps over the lazy {}'.format('fox','dog'))
print('the quick brown {1} jumps over the lazy {0}'.format('dog','fox'))
print('the quick brown {jumper} jumps over the lazy {jumpee}'.format(jumpee='dog',jumper='fox'))

D = {'a':'fox','b':'dog'}
print("\nthe quick brown {a} jumps over the lazy {b}".format(**D))
    # the double asterisk is unpacking dictionary D as if it were keyword arguments to the format function !
```
